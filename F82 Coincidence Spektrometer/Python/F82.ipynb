## Second part: $\gamma$-spectrum for different elements

We need to determine a calibration line for with our results of the measurement for  $^{60}$CO, $^{137}$Cs, $^{133}$Ba, $^{54}$Mn and $^{22}$Na.

```python
#Looking for the peaks in our measured spectra. Order: CO, CO, Cs, Mn, Na
channels = np.array([976, 1107, 557, 704, 1054])
#channels_err = 
energies = np.array([1.17323, 1.33248, 0.6616, 0.8353, 1.2746]) #MeV
#energies_err = 


#Plotting the result
plt.plot(channels[:1],energies[:1],color = 'darkred',label = r'$^{60}$Co', marker = 'o')
plt.plot(channels[2],energies[2],color = 'darkblue',label = r'$^{137}$Cs', marker = 'o')
plt.plot(channels[3],energies[3],color = 'darkgreen',label = r'$^{54}$Mn', marker = 'o')
plt.plot(channels[4],energies[4],color = 'darkorange',label = r'$^{22}$Na', marker = 'o')
plt.title('Calibration of the channels with theoretical energy values', size = 20)
plt.xlabel('Channel', size = 18)
plt.ylabel('Energy $E $ [MeV]', size = 18)

plt.legend(frameon=True,fontsize='large', borderpad=1.5, borderaxespad=1.5,loc='best')
plt.grid(ls='dotted')

#Saving the result
plt.tight_layout()
plt.savefig('Plots/calibration.pdf', format='PDF')
```
